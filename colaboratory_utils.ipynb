## Downloading and executing a file

````
import colaboratory_utils as colab_utils

fname = 'circle_polygon.py'
url = 'https://raw.githubusercontent.com/PacktPublishing/Mastering-OpenCV-4-with-Python/master/Chapter04/02-exercices/circle_polygon.py'

colab_utils.download_and_execute_file(fname, url, params= "", execute=True, show_content=True)
````

## Showing severl images in the same figure

````
!wget https://raw.githubusercontent.com/opencv/opencv/master/samples/data/HappyFish.jpg
````

````
import cv2
import matplotlib.pyplot as plt
import colaboratory_utils as colab_utils

img_bgr = cv2.imread("/content/HappyFish.jpg")
print("Loaded image with shape: '{}'".format(img_bgr.shape))
img_gray = cv2.imread("/content/HappyFish.jpg", cv2.IMREAD_GRAYSCALE)
print("Loaded image with shape: '{}'".format(img_gray.shape))

# Generate six new images to show.
img_bgr_flip_ud = cv2.flip(img_bgr, 0)
img_bgr_flip_lr = cv2.flip(img_bgr, 1)
img_bgr_flip_ud_lr = cv2.flip(img_bgr, -1)

img_gray_flip_ud = cv2.flip(img_gray, 0)
img_gray_flip_lr = cv2.flip(img_gray, 1)
img_gray_flip_ud_lr = cv2.flip(img_gray, -1)


# Create the dimensions of the figure and set title:
plt.figure(figsize=(12, 7))
plt.suptitle("Testing visualization (and flipping)", fontsize=14, fontweight='bold')

colab_utils.show_img_plt(img_bgr_flip_ud, title='flip ud', n_rows=2, n_cols=3, pos=1)
colab_utils.show_img_plt(img_bgr_flip_lr, title='flip lr', n_rows=2, n_cols=3, pos=2)
colab_utils.show_img_plt(img_bgr_flip_ud_lr, title='flip ud lr', n_rows=2, n_cols=3, pos=3)

colab_utils.show_img_plt(img_gray_flip_ud, is_color=False, title='flip ud', n_rows=2, n_cols=3, pos=4)
colab_utils.show_img_plt(img_gray_flip_lr, is_color=False, title='flip lr', n_rows=2, n_cols=3, pos=5)
colab_utils.show_img_plt(img_gray_flip_ud_lr, is_color=False, title='flip ud lr', n_rows=2, n_cols=3, pos=6)

# Show the created image:
plt.show()
````

## Taking video from the webcam

````
total_iter = 10

vid = colab_utils.videoGrabber(showVideo=False)

try:
  n_iter = 0
  while n_iter <= total_iter:
    n_iter += 1
    image_np = np.array(vid(0))
    colab_utils.imshow(image_np)
finally:
  vid(stop=True)
````
