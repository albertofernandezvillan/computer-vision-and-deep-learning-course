# Summary

In this notebook, we are going to see how to draw text and symbols. For drawing text the function `cv2.putText() `is used, and for drawing symbols `cv2.drawMarker()` is used.

# Drawing text in the image

To draw text in the image, the function `cv2.putText()` can be used. The function `cv2.putText()` renders the specified text string in the image. The signature of this function is:
```
img = cv2.putText(img, text, org, fontFace, fontScale, color[, thickness[, lineType[, bottomLeftOrigin]]])
```
Only a subset of [Hershey fonts](https://en.wikipedia.org/wiki/Hershey_fonts) are available. Parameter `fontFace` can take one of the following values:

```
FONT_HERSHEY_SIMPLEX
FONT_HERSHEY_PLAIN
FONT_HERSHEY_DUPLEX
FONT_HERSHEY_COMPLEX
FONT_HERSHEY_TRIPLEX
FONT_HERSHEY_COMPLEX_SMALL
FONT_HERSHEY_SCRIPT_SIMPLEX
FONT_HERSHEY_SCRIPT_COMPLEX
```

`fontScale`	defines the font scale factor that is multiplied by the font-specific base size.

We first define a function, which creates the canvas to draw:` 400x400` pixels, `3` channels, `uint8` (8-bit unsigned integers). We set the background to black using `np.zeros()`:

````
import numpy as np

def create_canvas():
  canvas = np.zeros((400, 400, 3), dtype="uint8")
  return canvas
````

````
import cv2
from matplotlib import pyplot as plt

image = create_canvas()

font = cv2.FONT_HERSHEY_TRIPLEX
cv2.putText(image, 'OpenCV', (5, 200), font, 3, (255,255,255), 1, cv2.LINE_AA)

plt.imshow(image[:, :, ::-1])
````

# Drawing symbols (markers)

OpenCV provides us with the `cv2.drawMarker()` function that allows us to draw one of the selected markers, we must indicate its position, the marker, its color and optionally the size of it. The available markers are:
```
MARKER_CROSS
MARKER_TILTED_CROSS
MARKER_STAR
MARKER_DIAMOND
MARKER_SQUARE
MARKER_TRIANGLE_UP
MARKER_TRIANGLE_DOWN
```




Therefore to draw the `MARKER_STAR` marker, we can do the following:

````
import cv2
from matplotlib import pyplot as plt

image = create_canvas()

marker = cv2.MARKER_STAR
print("Marker MARKER_STAR is equals to '{}'".format(marker))

cv2.drawMarker(image, (200,200), (0,255,0), marker, 200)

plt.imshow(image[:, :, ::-1])
````

Based on previous example, we can show all markers. See next piece of code to see how.

````
import cv2
from matplotlib import pyplot as plt

image = create_canvas()

for marker in range(0, 7):
  print("Drawing marker '{}'".format(marker))
  offset = 50 * marker
  cv2.drawMarker(image, (50 + offset, 50 + offset), (0,255,0), marker, 40, 3)

plt.imshow(image[:, :, ::-1])
````
