The second step is to load the image using OpenCV function `cv2.imread()`, which can be used to load the image from disk. The only required argument is the path of the image.

Remember that OpenCV loads the image in BGR format (rather than RGB format). RGB format is used by other Python packages, so we keep this in mind.

Optional argument can be used to indicate OpenCV how to load the image. For example, if we want to load the image in grayscale (`cv2.IMREAD_GRAYSCALE` value should be used as second parameter).

Do not forget to `import cv2` package (already installed in Colab)

```
# Import required packages:
import cv2

# Show imported version
print("OpenCV version:'{}'".format(cv2.__version__))
```

```
IMG_NAME = 'sample.png'

# Load the image using cv2.imread():
bgr_image = cv2.imread(IMG_NAME)

# Show the dimensions (and number of channels - if color image):
print('Shape of the loaded image: {}'.format(bgr_image.shape))
```

```
# Load the same image in grayscale:
gray_image = cv2.imread(IMG_NAME, cv2.IMREAD_GRAYSCALE)

# Show the dimensions (and number of channels - if color image):
print('Shape of the loaded image: {}'.format(gray_image.shape))
```

You can use the dimensions of the shape attribute to know if an image is loaded in grayscale mode or in color mode. # Some basic image processing
In this example, we are going to see how to apply some (basic) image processing algorithms using sliders, which is a nice way of setting the right value for the arguments used in your functions.

We only have to set the title and the parameters.

Python OpenCV package provides different ways for image smoothing also called blurring. In this example, we are going to see how to smooth images using `cv2.blur()`.

```
# We import matplotlib to show the resulting image 
%matplotlib inline
from matplotlib import pyplot as plt
import matplotlib

# Show imported version
print("matplotlib version:'{}'".format(matplotlib.__version__))
```

We define the function `blur_image`, which will be called later

```
def blur_image(src, kernel_size):
  result = cv2.blur(src,(kernel_size,kernel_size))
  return result
```

At this point, we can define the sliders, call the previous function, and showing the results using matplotlib capabilities

```
#@title Parameters for blurring { run: "auto" }
kernel_size_bgr = 13 #@param {type:"slider", min:1, max:20, step:1}
kernel_size_gray = 2 #@param {type:"slider", min:1, max:20, step:1}

dst_rgb = blur_image(bgr_image, kernel_size_bgr)
dst_gray = blur_image(gray_image, kernel_size_gray)

plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.imshow(dst_rgb[:,:,::-1])
plt.subplot(1, 2, 2)
plt.imshow(dst_gray, cmap='gray')
plt.show()
```
